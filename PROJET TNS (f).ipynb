```python
import numpy as np
import matplotlib.pyplot as plt

# Définir deux signaux numériques
signal1 = [1, 2, 3, 4, 5]  # Premier signal (échantillons)
signal2 = [2, 1, 0, -1, -2]  # Deuxième signal (échantillons)

# Effectuer la convolution des deux signaux
# La convolution combine les deux signaux pour produire un signal de sortie
# Elle est utilisée pour analyser comment un signal est modifié par un autre
convoluted_signal = np.convolve(signal1, signal2, mode='full')  # Mode 'full' produit la convolution complète

# Générer les indices temporels pour chaque signal
t_signal1 = np.arange(len(signal1))  # Indices pour le premier signal
t_signal2 = np.arange(len(signal2))  # Indices pour le second signal
t_convoluted = np.arange(len(convoluted_signal))  # Indices pour le signal convolué

# Tracer les signaux
plt.figure(figsize=(10, 6))  # Taille de la figure

# Tracer le Signal 1
plt.subplot(3, 1, 1)  # Premier sous-graphe
plt.stem(t_signal1, signal1)  # Tracé discret avec des tiges
plt.title("Signal 1")  # Titre du graphe
plt.xlabel("Temps (n)")  # Étiquette de l'axe des abscisses
plt.ylabel("Amplitude")  # Étiquette de l'axe des ordonnées

# Tracer le Signal 2
plt.subplot(3, 1, 2)  # Deuxième sous-graphe
plt.stem(t_signal2, signal2)  # Tracé discret avec des tiges
plt.title("Signal 2")  # Titre du graphe
plt.xlabel("Temps (n)")  # Étiquette de l'axe des abscisses
plt.ylabel("Amplitude")  # Étiquette de l'axe des ordonnées

# Tracer le Signal convolué
plt.subplot(3, 1, 3)  # Troisième sous-graphe
plt.stem(t_convoluted, convoluted_signal)  # Tracé discret avec des tiges
plt.title("Signal Convolué")  # Titre du graphe
plt.xlabel("Temps (n)")  # Étiquette de l'axe des abscisses
plt.ylabel("Amplitude")  # Étiquette de l'axe des ordonnées

# Ajuster l'espacement entre les sous-graphes pour éviter les chevauchements
plt.tight_layout()  

# Afficher les graphiques
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définir deux signaux numériques
signal1 = [1, 2, 3, 4, 5]  # Premier signal discret
signal2 = [2, 1, 0, -1, -2]  # Deuxième signal discret

# Effectuer la convolution des deux signaux
# La convolution calcule la somme croisée des échantillons des deux signaux
convoluted_signal = np.convolve(signal1, signal2, mode='full')  # Mode 'full' retourne la convolution complète

# Calculer la Transformée de Fourier rapide (FFT) du signal convolué
fft_result = np.fft.fft(convoluted_signal)  # La FFT transforme le signal convolué en domaine fréquentiel
frequencies = np.fft.fftfreq(len(convoluted_signal))  # Calcul des fréquences correspondantes

# Calculer la magnitude (ou spectre d'amplitude)
magnitude = np.abs(fft_result)  # La magnitude donne l'amplitude de chaque composante fréquentielle

# Tracer les graphiques
plt.figure(figsize=(12, 6))  # Définir la taille de la figure

# Graphique 1 : Signal convolué
plt.subplot(2, 1, 1)  # Premier sous-graphe
plt.stem(np.arange(len(convoluted_signal)), convoluted_signal, basefmt=" ")  # Tracé discret du signal convolué
plt.title("Signal Convolué")  # Titre du graphe
plt.xlabel("Temps (n)")  # Étiquette pour l'axe des abscisses
plt.ylabel("Amplitude")  # Étiquette pour l'axe des ordonnées
plt.grid()  # Afficher une grille pour faciliter la lecture

# Graphique 2 : Spectre de fréquence (FFT)
plt.subplot(2, 1, 2)  # Deuxième sous-graphe
plt.stem(frequencies, magnitude, basefmt=" ")  # Tracé discret des fréquences et amplitudes
plt.title("Spectre de Fréquence (FFT)")  # Titre du graphe
plt.xlabel("Fréquence")  # Étiquette pour l'axe des abscisses
plt.ylabel("Amplitude")  # Étiquette pour l'axe des ordonnées
plt.grid()  # Afficher une grille

# Ajuster l'affichage des sous-graphes pour éviter le chevauchement
plt.tight_layout()

# Afficher les graphiques
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définir deux signaux numériques sous forme de listes
signal1 = [1, 2, 3, 4, 5]
signal2 = [2, 1, 0, -1, -2]

# Effectuer la convolution des deux signaux
# La convolution est une opération qui combine deux signaux pour en produire un troisième
convoluted_signal = np.convolve(signal1, signal2, mode='full')

# Créer un filtre idéal passe-bas
# Ici, un filtre passe-bas est simulé par une moyenne mobile de longueur 7
filter_length = 7
ideal_filter = np.ones(filter_length) / filter_length  # Moyenne mobile

# Appliquer le filtre idéal au signal convolué
# La convolution du signal convolué avec le filtre idéal permet de lisser le signal
filtered_signal_ideal = np.convolve(convoluted_signal, ideal_filter, mode='same')

# Générer les indices temporels pour les signaux
# Ces indices seront utilisés pour l'axe des abscisses lors du tracé des signaux
t_convoluted = np.arange(len(convoluted_signal))  # Axe temporel pour le signal convolué
t_filtered = np.arange(len(filtered_signal_ideal))  # Axe temporel pour le signal filtré

# Tracer les signaux
plt.figure(figsize=(10, 8))

# Tracer le signal convolué
plt.subplot(4, 1, 3)
plt.stem(t_convoluted, convoluted_signal, 'g', label="Signal Convolué")
plt.title("Signal Convolué")
plt.xlabel("Temps (n)")
plt.ylabel("Amplitude")
plt.grid(True)

# Tracer le signal filtré (après application du filtre idéal passe-bas)
plt.subplot(4, 1, 4)
plt.stem(t_filtered, filtered_signal_ideal, 'r', label="Filtre Idéal Passe-Bas")
plt.title("Signal Filtré (Filtre Idéal Passe-Bas)")
plt.xlabel("Temps (n)")
plt.ylabel("Amplitude")
plt.grid(True)

# Ajuster la disposition des sous-graphiques pour éviter le chevauchement
plt.tight_layout()

# Afficher les graphiques
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Définir deux signaux numériques sous forme de listes
signal1 = [1, 2, 3, 4, 5]
signal2 = [2, 1, 0, -1, -2]

# Effectuer la convolution des deux signaux
# La convolution est effectuée en mode 'full' pour obtenir la sortie complète
convoluted_signal = np.convolve(signal1, signal2, mode='full')

# Paramètres du filtre IIR (Infinite Impulse Response)
cutoff_freq = 0.3  # Fréquence de coupure normalisée (entre 0 et 0.5, 0.5 étant la fréquence Nyquist)
filter_order = 4   # Ordre du filtre

# Conception du filtre IIR passe-bas de type Butterworth
# La fonction 'butter' retourne les coefficients du filtre (b, a)
b, a = butter(filter_order, cutoff_freq, btype='low', analog=False)

# Appliquer le filtre IIR au signal convolué
# La fonction 'lfilter' applique le filtre défini par les coefficients b et a au signal
filtered_signal_iir = lfilter(b, a, convoluted_signal)

# Générer les indices pour les signaux
# 't_convoluted' et 't_filtered' représentent les indices temporels des signaux
t_convoluted = np.arange(len(convoluted_signal))
t_filtered = np.arange(len(filtered_signal_iir))

# Tracer les signaux
plt.figure(figsize=(10, 6))

# Tracer le signal convolué
plt.subplot(2, 1, 1)
plt.stem(t_convoluted, convoluted_signal, basefmt=" ", linefmt='b', markerfmt='bo', label="Signal Convolué")
plt.title("Signal Convolué")
plt.xlabel("Temps (n)")
plt.ylabel("Amplitude")
plt.legend()

# Tracer le signal filtré (après application du filtre IIR)
plt.subplot(2, 1, 2)
plt.stem(t_filtered, filtered_signal_iir, basefmt=" ", linefmt='g', markerfmt='go', label="Signal Filtré (IIR)")
plt.title("Signal Filtré (Filtre IIR)")
plt.xlabel("Temps (n)")
plt.ylabel("Amplitude")
plt.legend()

# Ajuster la disposition des sous-graphes pour éviter le chevauchement
plt.tight_layout()

# Afficher les graphiques
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, lfilter

# Définir deux signaux numériques
signal1 = [1, 2, 3, 4, 5]  # Premier signal
signal2 = [2, 1, 0, -1, -2]  # Deuxième signal

# Effectuer la convolution des deux signaux
convoluted_signal = np.convolve(signal1, signal2, mode='full')

# Paramètres du filtre FIR
filter_order = 7  # Ordre du filtre (nombre de coefficients - 1)
cutoff_freq = 0.3  # Fréquence de coupure normalisée (0 à 0.5 pour la fréquence Nyquist)

# Conception du filtre FIR passe-bas avec une fenêtre de Hamming
fir_filter = firwin(filter_order + 1, cutoff_freq, window='hamming')

# Appliquer le filtre FIR au signal convolué
filtered_signal_real = lfilter(fir_filter, 1.0, convoluted_signal)

# Générer les indices pour les signaux
t_convoluted = np.arange(len(convoluted_signal))  # Indices temporels pour le signal convolué
t_filtered = np.arange(len(filtered_signal_real))  # Indices temporels pour le signal filtré

# Tracer les signaux
plt.figure(figsize=(10, 6))

# Signal convolué
plt.subplot(2, 1, 1)
plt.stem(t_convoluted, convoluted_signal, basefmt=" ")
plt.title("Signal Convolué")
plt.xlabel("Temps (n)")
plt.ylabel("Amplitude")

# Signal filtré (Filtre FIR)
plt.subplot(2, 1, 2)
plt.stem(t_filtered, filtered_signal_real, basefmt=" ", linefmt='r')
plt.title("Signal Filtré (Filtre FIR Numérique Réel)")
plt.xlabel("Temps (n)")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
